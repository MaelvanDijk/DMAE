# First graphs

Verbruiken auto's met een grotere motor ook meer benzine? Dit is de eerste vraag die we gaan beantwoorden met ggplot2. Hiervoor gebruiken we de *mpg* dataset in de **ggplot2** library.

de eerster variabele waar we naar kijken:
* dspl: motor inhoud in liters
* hwy: brandstof efficientie van de auto, lager is minder efficient

````r
ggplot2::mpg %>% # load dataset
ggplot(aes(x= displ, y = hwy)) + # make canvas
geom_point() # populate canvas with points
````

````r
mpg %>%
ggplot() +
geom_point(mapping= aes(x=displ, y= hwy))
````

## Graphing template
De code die hierboven wordt gebruikt voor het maken van een grafiek volgt een bepaald template namelijk:
```
ggplot(data = <DATA>)+
    <GEOM_FUNCTION>(mapping= aes(<MAPPINGS>))
```

ggplot is het canvas en de geom functie tekent hier een object (scatterplot, barchart etc.) op.

````r
options(repr.plot.width=8, repr.plot.height=8)

mpg

?mpg

mpg %>%
    ggplot(aes(class, drv)) +
    geom_point()
````

## Aestetics mapping

````r
mpg %>%
    ggplot(
        aes(
            displ, # x-as
            hwy, # y-as
            color= class # kleur
            )
        ) +
    geom_point()
````

````r
mpg %>%
    ggplot(
        aes(
            displ, # x-as
            hwy, # y-as
            size= class # genereert warning, size verwacht een volgordelijkheid
            )
        ) +
    geom_point()

mpg %>%
    ggplot(
        aes(
            displ, # x-as
            hwy, # y-as
            alpha= class # genereert warning, alpha verwacht volgordelijkheid
            )
        ) +
    geom_point()

mpg %>%
    ggplot(
        aes(
            displ, # x-as
            hwy, # y-as
            shape= class # mapped max 6 variabele
            )
        ) +
    geom_point()
````

Je kan naast ggplot gebruiken voor de aestetics ook de aestetics meegeven in de `geom()` functie

````r
mpg %>%
    ggplot() +
    geom_point(mapping= aes(displ, hwy), color= "blue")
````

````r
mpg %>%
    ggplot() +
    geom_point(mapping= aes(
        x= displ,
        y= hwy,
        color= "blue" # dit zal niet werken. De kleur moet bepaald worden buiten aes()
        )
        )
````

````r
head(mpg, 5) # laat een klein stukje van de tabel zien
str(mpg) # get data type of cols str() --> structure
````

````r
mpg %>%
    ggplot() +
    geom_point(mapping= aes(
        x= displ,
        y= hwy,
        color= year,
        size= cty
        # shape = cyl --> shape cannot be continuous
        )
        )

mpg %>%
    ggplot() +
    geom_point(mapping= aes(
        x= displ,
        y= hwy,
        color= year, #can map the same variabele twice
        size= year
        )
        )

mpg %>%
    ggplot() +
    geom_point(mapping= aes(
        x= displ,
        y= hwy,
        stroke= 5, # lijn dikte van de punten
        color= year
        )
        )

mpg %>%
    ggplot() +
    geom_point(mapping= aes(
        x= displ,
        y= hwy,
        color= (displ < 5) # Makes a boolean of every point 
        )
        )
````

## Facets

subplots maken om individuele grafieken duidelijker weer te geven. Er zijn twee smaken:
* facet_wrap() = subplots op basis van 1 variabele
* facet_grid() = subplots op basis van 2 variabele

````r
mpg %>%
    ggplot() +
    geom_point(mapping= aes(
        x= displ,
        y= hwy
        ))+
    facet_wrap( # creates subplots
        ~class, # ~ == "formula" a datastructure in R
        nrow= 2 # number of rows
    )

mpg %>%
    ggplot() +
    geom_point(mapping= aes(
        x= displ,
        y= hwy
        ))+
    facet_grid( # creates subplots
        drv ~ cyl,#facet grid takes two variables
    )

mpg %>%
    ggplot() +
    geom_point(mapping= aes(
        x= displ,
        y= hwy
        ))+
    facet_grid( # creates subplots
        . ~ cyl,#kan gebruikt worden om kolommen te maken
    )

mpg %>%
    ggplot() +
    geom_point(mapping= aes(
        x= displ,
        y= hwy
        ))+
    facet_grid( # creates subplots
        drv ~ . ,#kan gebruikt worden om rijen te maken
    )
````

````r
mpg %>%
    ggplot() +
    geom_point(mapping= aes(
        x= displ,
        y= hwy
        ))+
    facet_wrap( # creates subplots
        ~year, # kan op continue variabelen
    )
````

````r
mpg %>%
    ggplot() +
    geom_point(mapping= aes(
        x= drv,
        y= cyl
        ))
    
````

## Geometric Objects
`geom_` controleert het type object (bar, line, point etc)

````r
mpg %>%
    ggplot() +
    geom_point( # scatterplot
        mapping= aes(
        x= displ,
        y= hwy,
        color= drv #maak de onderverdeling duidelijker met kleur
        )) +
    geom_smooth( # glad gestreken scatterplot (y gegeven x)
        mapping= aes(
        x= displ,
        y= hwy,
        color= drv, #hier ook kleur om te relateren
        linetype= drv # lijn type als 3e variabele
        ))

mpg %>%
    ggplot(aes(x= displ, y= hwy)) + #ggplot(aes()) mapped variabelen globaal
    geom_point()+ # Hier hoeven nu geen variabele meegegeven te worden
    geom_smooth() # hier ook niet
````
