## Graphing template
De code die hierboven wordt gebruikt voor het maken van een grafiek volgt een bepaald template namelijk:
```
ggplot(data = <DATA>)+
    <GEOM_FUNCTION>(mapping= aes(<MAPPINGS>))
```

ggplot is het canvas en de geom functie tekent hier een object (scatterplot, barchart etc.) op.

````r
options(repr.plot.width=8, repr.plot.height=8)

mpg

?mpg

mpg %>%
    ggplot(aes(class, drv)) +
    geom_point()
````

## Facets

subplots maken om individuele grafieken duidelijker weer te geven. Er zijn twee smaken:
* facet_wrap() = subplots op basis van 1 variabele
* facet_grid() = subplots op basis van 2 variabele

````r
mpg %>%
    ggplot() +
    geom_point(mapping= aes(
        x= displ,
        y= hwy
        ))+
    facet_wrap( # creates subplots
        ~class, # ~ == "formula" a datastructure in R
        nrow= 2 # number of rows
    )

mpg %>%
    ggplot() +
    geom_point(mapping= aes(
        x= displ,
        y= hwy
        ))+
    facet_grid( # creates subplots
        drv ~ cyl,#facet grid takes two variables
    )

mpg %>%
    ggplot() +
    geom_point(mapping= aes(
        x= displ,
        y= hwy
        ))+
    facet_grid( # creates subplots
        . ~ cyl,#kan gebruikt worden om kolommen te maken
    )

mpg %>%
    ggplot() +
    geom_point(mapping= aes(
        x= displ,
        y= hwy
        ))+
    facet_grid( # creates subplots
        drv ~ . ,#kan gebruikt worden om rijen te maken
    )
````

````r
mpg %>%
    ggplot() +
    geom_point(mapping= aes(
        x= displ,
        y= hwy
        ))+
    facet_wrap( # creates subplots
        ~year, # kan op continue variabelen
    )
````

````r
mpg %>%
    ggplot() +
    geom_point(mapping= aes(
        x= drv,
        y= cyl
        ))
    
````

## Statistical transformation

Datasets hebben vaak geen variabele "count". In ggplot kunnen grafieken verschillende statistische transformaties direct uitvoeren zoals:

* binnen: bar charts, histograms en frequency polygons. De data "binnen" (in een bakje stoppen) en de counts hier uit opmaken.
* "smoothers" vallen (zoals geom_smooth) fitten a model op de data en doen op deze manier direct een statistische transformatie
* boxplots, deze grafiek voert een aantal samenvattende statistieken uit.

onder de motorkap gebruiken deze grafieken in ggplot statistische functies die tot de gewenste statistische transformaties leiden.

````r
diamonds <- ggplot2::diamonds

# barplot example
diamonds %>%
    ggplot() +
    geom_bar(mapping= aes(x= cut))

# same example but with stat_count()
diamonds %>%
    ggplot() +
    stat_count(mapping= aes(x=cut))
````

De standaard statistische transformatie van `geom_` kan worden aangepast met het argument stat `geom_bar(mapping= aes(), stat= "identity"` hier zorgt identity er voor dat de kolom waarde wordt getoond. Zo kunnen ook andere stats worden aangepast (ook in de `aes()` argument).

````r
# laat bars als proportie van het geheel zien
diamonds %>%
    ggplot() +
    geom_bar(mapping= aes(x= cut, y= ..prop.., group= 1))

# ook kan specifieke statistieken worden aangeroepen
diamonds %>%
    ggplot() +
    stat_summary( # stats voor geom_pointrange 
        mapping = aes(x = cut, y = depth),
        fun.ymax = max, # stelt max van het bereik in
        fun.ymin = min, # stelt min van het bereik in
        fun.y = median # stelt median van het bereik in
    )
````

###  Excercise
1. wat doet `geom_col()` anders dan `geom_bar()`? --> de hoogte van de bar geven de daadwerkelijke variabelen weer (in dit geval waarschijnlijk een som van de variabele).
2. welke variabele berekend `stat_smooth()`? --> voorspelde x of y waarde, de min/max hiervan en de standard error.
3. waarom moet bij ..prop.. in de barchart hierboven ook een group= 1 mee worden gegeven? --> group= 1 zorgt dat de bars ten opzichten van elkaar worden berekend en niet elke bar = 100%

````r
# geom_col 
diamonds %>%
    ggplot() +
    geom_col(
        mapping = aes(x = cut, y =depth)
    )

# stat_smooth
diamonds %>%
    ggplot()+
    geom_smooth(
        mapping = aes(x = carat, y = depth)
    )

# ..prop..
diamonds %>%
    ggplot() +
    geom_bar(mapping= aes(x= cut, y= ..prop..))

diamonds %>%
    ggplot() +
    geom_bar(mapping= aes(x= cut, y= ..prop.., fill = color))
````

## Position Adjustments
binnen plots kan je zowel `color=` als `fill=` gebruiken.
In de laatse bar chart worden de kleuren gestapeld door de *stacking argument*. 

````r
# color (alleen outline is gekleurd)
diamonds %>%
    ggplot() +
    geom_bar(mapping= aes(x= cut, color= cut))

# fill volledige bar
diamonds %>%
    ggplot() +
    geom_bar(mapping= aes(x= cut, fill= cut))

# fill volledige bar met verschillende waardes
diamonds %>%
    ggplot() +
    geom_bar(mapping= aes(x= cut, fill= clarity))
````

Om de posities te bepalen kunnen verschillende `position= ` argumenten worden gebruikt, zoals: `identity` `fill` `dodge`

````r
# identity --> niet handig voor bars want de data punten worden
# gemapped precies op de plek waar ze horen (dit zorgt bij bars voor overlap)
diamonds %>%
    ggplot(mapping= aes(x= cut, fill= clarity)) +
    geom_bar(
        alpha = 1/5,
        position= "identity"
    )
    
diamonds %>%
    ggplot(mapping= aes(x= cut, color= clarity)) +
    geom_bar(
        fill = NA,
        position= "identity"
    ) 

# fill stapelt de bars tot 100% binnen de eigen groep
# fill volledige bar
diamonds %>%
    ggplot() +
    geom_bar(
        mapping= aes(x= cut, fill= clarity),
        position= "fill"
        )

# dodge zet alle subgroepen naast elkaar
diamonds %>%
    ggplot() +
    geom_bar(
        mapping= aes(x= cut, fill= clarity),
        position = "dodge"
    )
````

in het eerdere voorbeeld van de auto scatterplot hebben we te maken met overplotting, niet alle datapunten zijn zichtbaar omdat ze overlappen. hiervoor kan je `jitter` gebruiken. Dit voegt een beetje randomness aan de punten toe om op een grafiek net iets meer te kunnen zien.

````r
# voeg jitter toe
mpg %>%
    ggplot() +
    geom_point(
        mapping= aes(x=displ, y=hwy),
        position= "jitter"
    )

# kan ook via geom_jitter
mpg %>%
    ggplot() +
    geom_jitter(
        mapping= aes(x=cty, y=hwy),
        width = 5, #controleert de hoeveelheid jitter in de breedte
        height = 5 # controleert de hoeveelheid jitter in de hoogte
    )
````

## coordinate systems

normaal gesproken plot ggplot2 in x,y ofwel carthesian coordinaten. Soms kan het handig zijn deze te controleren bijvoorbeeld wanneer je de coordinaten wilt draaien. Daarnaast kan je geospatial data plotten (dit wordt verder niet behandeld) en polar coordinaten gebruiken (ronde vormen)

````r
# normale plot
mpg %>%
    ggplot() +
    geom_point(
        mapping= aes(x=displ, y=hwy)
    )

# geflipte plot
mpg %>%
    ggplot() +
    geom_point(
        mapping= aes(x=displ, y=hwy)
    ) +
    coord_flip()

# polar coordinaten
diamonds %>%
    ggplot()+
    geom_bar(
        mapping = aes(x = cut, fill= cut),
        show.legend = FALSE,
        width = 1
    ) +
    theme(aspect.ratio = 1) +
    labs(x = NULL, y = NULL) + # past text van legenda, titels en assen aan
    coord_flip() +
    coord_polar() # maakt een circel diagram
````
