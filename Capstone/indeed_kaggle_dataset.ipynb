# indeed kaggle data

```r
library(readr)
library(tidyverse)

indeed <- read_csv(".\\data\\indeed_job_dataset.csv", show_col_types = FALSE)
head(indeed)

# indeed$Skill <- as.factor(indeed$Skill)
```

```r
indeed <- indeed %>% 
  mutate(Skill = strsplit(as.character(Skill), ",")) %>%
  unnest(Skill)
```

```r
head(indeed, 5)
```

```r
indeed$Skill <- gsub('[[]', '', indeed$Skill)
indeed$Skill <- gsub('[]]', '', indeed$Skill)
indeed$Skill <- gsub("[']", '', indeed$Skill)
head(indeed)
names(indeed)
indeed$Date_Since_Posted
```

```r
# Store most sought after skills in list
s <- table(indeed$Skill)
s <- s[s > 200] %>% sort(decreasing = TRUE)

# Make plots wider 
options(repr.plot.width=15, repr.plot.height=12)


indeed %>%
    dplyr::filter(Skill %in% names(s)) %>% # filter df based on most sought skill
    ggplot(
        aes(x = reorder(
            Skill, Skill, function(x) + length(x)), # decreasing order skill
            fill= Job_Type
            )) +
    geom_bar() +
    theme(axis.text.y = element_text(
        face= 'bold',
        size= 14)) + # Rotate x-labels and change font
        coord_flip() 
        # facet_wrap(~Job_Type)
```

```r
skills <- table(indeed$Skill)
skill_names <- names(skills)
trimmed_skill_names<- trimws(skill_names, "left") %>% unique()
```

```r
trimmed_skill_names
write.table(trimmed_skill_names,"trimmed_skill_names.txt",sep="\t",row.names=FALSE, append = FALSE)
```
